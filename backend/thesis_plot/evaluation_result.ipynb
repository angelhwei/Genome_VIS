```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Task time
data1 = [139, 151, 121, 136]
data2 = [225, 220, 177, 226]
data3 = [182, 144, 185, 152]

data = [data1, data2, data3]
labels = ['Chromosome/Scaffold View', 'Sequence View', 'Gene Expression View']
```

```python
# 绘制箱形图
meanprops = dict(marker='D', markeredgecolor='c', markerfacecolor='c')

plt.boxplot(data, patch_artist=True, showfliers=True, showmeans=True,
            meanprops=meanprops, boxprops=dict(facecolor='w', color='black'))
plt.title('Task Time')
plt.ylabel('Time (sec)')

# 设置 x 轴的标签
plt.xticks([1, 2, 3], labels)

# 计算并显示每组数据的平均值和中位数
# for i in range(len(data)):
#     mean = np.mean(data[i])
#     median = np.median(data[i])
#     plt.text(i+1, mean, f'Mean: {mean:.2f}', ha='center', va='bottom', fontsize=8)
#     plt.text(i+1, median, f'Median: {median:.2f}', ha='center', va='top', fontsize=8)
```

```python
# Task accuracy
data1 = [100, 100, 100, 100]
data2 = [58, 75, 86, 86]
data3 = [100, 100, 67, 50]

accuracy_data = [data1, data2, data3]
labels = ['Chromosome/Scaffold View', 'Sequence View', 'Gene Expression View']
```

```python
# 绘制箱形图
meanprops = dict(marker='D', markeredgecolor='m', markerfacecolor='m')

plt.boxplot(accuracy_data, patch_artist=True, showfliers=True, showmeans=True,
            meanprops=meanprops, boxprops=dict(facecolor='w', color='black'))
plt.title('Task Accuracy')
plt.ylabel('Accuracy (%)')

# 设置 y 轴的范围
plt.ylim(bottom=0)

# 设置 x 轴的标签
plt.xticks([1, 2, 3], labels)

# 计算并显示每组数据的平均值和中位数
# for i in range(len(accuracy_data)):
#     mean = np.mean(accuracy_data[i])
#     median = np.median(accuracy_data[i])
#     plt.text(i+1, mean, f'Mean: {mean:.2f}', ha='center', va='bottom', fontsize=8)
#     plt.text(i+1, median, f'Median: {median:.2f}', ha='center', va='top', fontsize=8)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 数据
labels = ['E5', 'E7', 'E8']
values1 = [5, 4, 5]
values2 = [5, 5, 4]

# 创建 x 的位置
x = np.arange(len(labels))

# 创建条形图
width = 0.35  # 条形的宽度
plt.bar(x - width/2, values1, width, label='Future System Utilization Potential', color='orange')
plt.bar(x + width/2, values2, width, label='Likelihood of Recommending the System', color='c')

# 添加标签，标题和图例
plt.ylabel('Scores')
plt.title('Figure Utilization & Recommendation')
plt.xticks(x, labels)
plt.legend(loc='lower right')

# 显示图形
plt.show()
```
